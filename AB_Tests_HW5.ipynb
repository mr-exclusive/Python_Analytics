$$ p2 - p1 \pm 1.96 \sqrt{ \frac{p1(1-p1)}{n1} + \frac{p2(1-p2)}{n2}} $$

> По формуле вычисляем доверительный интервал и смотрим, если 0 входить в этот интервал, то делаем заключение что статистически значимые отличия не были найдены между исследуемыми группами и проверяемые изменения в продукте не подтвердили гипотезу.

> Найдем доверительный интервал для разницы конверсий контрольной группы (1.05%) и тестовой группы (1.47%)

```python
p1 = 0.0105
p2 = 0.0147
n1 = n2 = 15550

p2 - p1 - 1.96 * (p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2) ** 0.5, p2 - p1 + 1.96 * (p1 * (1 - p1) / n1 + p2 * (1 - p2) / n2) ** 0.5
```

> 0 не входит в найденый доверительный интервал (0.17%, 0.67%), это значит что между группами ЕСТЬ статистически значимые различия.
